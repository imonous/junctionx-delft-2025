```python
import pandas as pd
```

```python
earnings = pd.read_csv('./data/uber_hackathon_v2_mock_data.xlsx - earnings_daily.csv', parse_dates=['date'])
weather  = pd.read_csv('./data/uber_hackathon_v2_mock_data.xlsx - weather_daily.csv', parse_dates=['date'])
```

```python
earnings.head(10)
```

```python
df = earnings.merge(weather, on=['city_id','date'], how='left')
df.head(10)
```

```python
rides_df = df[[
    "earner_id","city_id","date","week","weather",
    "trips_count","rides_distance_km","rides_duration_mins",
    "rides_gross_fare","rides_net_earnings","rides_tips"
]].copy()

rides_df.head()
```

```python
eats_df = df[[
    "earner_id","city_id","date","week","weather",
    "orders_count","eats_distance_km","eats_duration_mins",
    "eats_delivery_fee","eats_net_earnings","eats_tips"
]].copy()

eats_df.head()
```

```python
df["weather"].unique()
```

```python
rides_avg = rides_df.groupby("weather", as_index=False)["rides_net_earnings"].mean().rename(columns={"rides_net_earnings":"avg_rides_net"})
print(rides_avg)
```

```python
df["trips_count"].describe()
```

```python
active = df[df["trips_count"] > 0].copy()

# Per-trip earnings for each earner-day
active["eur_per_trip"] = active["rides_net_earnings"] / active["trips_count"]

per_trip_by_weather = (
    active.groupby("weather", as_index=False)["eur_per_trip"]
          .mean()
          .rename(columns={"eur_per_trip":"avg_eur_per_trip"})
          .sort_values("avg_eur_per_trip", ascending=False)
)
print(per_trip_by_weather)
```
