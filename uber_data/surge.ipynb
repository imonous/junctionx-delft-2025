```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/uber_hackathon_v2_mock_data.xlsx - surge_by_hour.csv")
df = df.sort_values(["city_id", "hour"])
df.tail(10)
```

```python
plt.figure(figsize=(10, 6))
for cid, g in df.groupby("city_id"):
    plt.plot(g["hour"], g["surge_multiplier"], marker="o", linewidth=2, label=f"City {cid}")

plt.title("Surge Multiplier by Hour for Each City")
plt.xlabel("Hour of Day (023)")
plt.ylabel("Surge Multiplier")
plt.xticks(range(0, 24))
plt.grid(True, alpha=0.3)
plt.legend(title="city_id", ncol=2)
plt.tight_layout()
plt.show()
```
